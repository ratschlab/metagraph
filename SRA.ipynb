```python
DIR = '/cluster/home/mikhaika/metagenome/benchmark_human_metagenome/nobackup/human_gut_sra'

filtered_df['RealSize'] = filtered_df.Run.apply(
    lambda x: os.path.getsize(DIR + '/' + x + '.sra')
)
```

```python
filtered_df[filtered_df.RealSize == 0].Run.to_csv('runs_human_gut_download_again.csv', index=False)
```

```python
fig, ax = plt.subplots(figsize=(15, 9))
filtered_df.plot.scatter('RealSize', 'Total Size, Mb', ax=ax)
ax.set_xscale('symlog')
ax.set_yscale('symlog')
ax.set_xlim([-1, 10**12])
ax.set_ylim([-1, 10**6])
ax.set_ylabel('Total Size, Mb', size=20)
ax.set_xlabel('Size on Disk, bytes', size=20)
for i in np.argsort(filtered_df['Total Size, Mb'] / filtered_df['RealSize'])[::-1][:5].values:
    ax.annotate(filtered_df.iloc[i].loc['Run'], xy=(filtered_df.iloc[i].loc['RealSize'],
                                                    filtered_df.iloc[i].loc['Total Size, Mb']))
plt.grid()
plt.show()
```

```python
sum(filtered_df.size_MB == 0)
```

```python
fig, ax = plt.subplots(figsize=(15, 9))
filtered_df.plot.scatter('RealSize', 'size_MB', ax=ax)
ax.set_xscale('symlog')
ax.set_yscale('symlog')
ax.set_xlim([-1, 10**12])
ax.set_ylim([-1, 10**6])
ax.set_ylabel('Expected size, Mb', size=20)
ax.set_xlabel('Size on Disk, bytes', size=20)
for i in np.argsort(filtered_df['size_MB'] / filtered_df['RealSize'])[::-1][:5].values:
    ax.annotate(filtered_df.iloc[i].loc['Run'], xy=(filtered_df.iloc[i].loc['RealSize'],
                                                    filtered_df.iloc[i].loc['size_MB']))
plt.grid()
plt.show()
```

```python
DIR = '/cluster/home/mikhaika/metagenome/benchmark_human_metagenome/nobackup/human_gut_sra'

filtered_df['FastqSize'] = filtered_df.Run.apply(
    lambda x: os.path.getsize(DIR + '/' + x + '.fastq.gz')
                if os.path.exists(DIR + '/' + x + '.fastq.gz') else 0
)
```
