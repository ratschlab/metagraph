### Metasub

Metasub dnaloc.ethz.ch databases are available on a remote host.

1. Add this config to `~/.ssh/config`
```
Host dna 
 User XXX
 HostName dnaloc.ethz.ch
 LocalForward 42623 127.0.0.1:42623
 LocalForward 42625 127.0.0.1:42625
 LocalForward 42627 127.0.0.1:42627
```
2. Connect to the server to setup port forwarding.
3. Now you can connect to metasub databases with this API

````python
from IPython.display import display
````

````python
metagraph = Client()

# metasub k=13
metagraph.add_graph('localhost', 42623, 'metasub k=13')
# metasub k=15
metagraph.add_graph('localhost', 42625, 'metasub k=15')
# metasub k=17
metagraph.add_graph('localhost', 42627, 'metasub k=17')

for i, table in metagraph.search('CAAATTCTTGTAAGTATTAAACATTGTATATGTATTTTGAA').items():
    print('\nDatabase {}'.format(i))
    display(table.head())
````
