```python
over_120 = open('/Users/dd/plants/over_120GB.ids').readlines()
eukaryota = set(line.strip() for line in open('/tmp/mostly_eukaryota').readlines())
id_to_size = dict((line.split(' ')[0], line.strip().split(' ')[1]) for line in over_120)

ncbi_data = open('/Users/dd/plants/plants.33090.2020.04.15.csv').readlines()
id_to_method = {}
for line in ncbi_data:
    els = line.split(',')  
    id_to_method[els[0]] = els[18]
    

transferred_ids = set(line.strip() for line in open('/Users/dd/Downloads/.metagraph_run37/succeed_transfer.id'))
transferred_ids.update((line.strip() for line in open('/Users/dd/Downloads/.metagraph_run39/succeed_transfer.id')))

transferred = 0
long_reads = 0
max_size = 0
for id, size in id_to_size.items():
    if id in transferred_ids:
        transferred += 1
        continue
    if id_to_method[id] == 'PACBIO_SMRT' or id_to_method[id] == 'OXFORD_NANOPORE':
        long_reads += 1
        continue
    if id not in eukaryota:
        continue
    sz = float(size)
    if sz > max_size:
        max_size = sz
    print(id)



print(f'Already transferred: {transferred}. Long reads: {long_reads}')
print(f'max size {max_size}')
```
