```python
%matplotlib inline
import sys
import scipy as sp
import os
import h5py
import glob
import re
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.pylab import cm 
import gromics.viz.axes as axs
```

```python
K = 31
basedir = '/cluster/work/grlab/projects/metagenome/data/metagut/parameterstudy/output_k%i_cleaned'% K
files = glob.glob(os.path.join(basedir, '*clean_s2.stats'))
fnames = sp.array([os.path.basename(_) for _ in files])
samples = sp.unique([_.split('.')[0] for _ in fnames])
plotdir = os.path.join(basedir, 'plots')
if not os.path.exists(plotdir):
    os.makedirs(plotdir)
```

```python
### collect all data data
all_data = []
for fname in files:
    ## graph stats
    all_data.append(sp.loadtxt(fname, dtype='float', delimiter='\t'))
    ## alignment stats
    fbase = os.path.basename(fname)
    fname_align = os.path.join(basedir + '_realign', re.sub('clean_s2', 'clean_s1.align', fbase))
    _tmp = sp.loadtxt(fname_align, dtype='float')
    all_data[-1] = sp.append(all_data[-1], _tmp[1] / sp.sum(_tmp))
all_data = sp.array(all_data)
```

```python
all_data.shape
```

```python
gs = gridspec.GridSpec(2, 3)
#colors = ['r', 'g', 'b', 'm']
colors = ['#d73027',
'#fc8d59',
'#fee08b',
'#d9ef8b',
'#91cf60',
'#1a9850']

tfs = sp.arange(6) + 1

for sample in samples:
    curr_data = []
    for tf in tfs:
        sidx = sp.where([_.startswith(sample) and 'tf_%i' % tf in _ for _ in fnames])[0]
        fbs = sp.array([int(_.split('.')[2].split('_')[1]) for _ in fnames[sidx]])
        ssidx = sp.argsort(fbs)
        fbs = fbs[ssidx]
        sidx = sidx[ssidx]
        curr_data.append(all_data[sidx])
    curr_data = sp.array(curr_data)
    
    fig = plt.figure(figsize=(15, 10), dpi=200)
    ### total counts
    ax = fig.add_subplot(gs[0, 0])
    for i in range(len(tfs)):
        ax.bar(sp.arange(curr_data.shape[1]) + 0.15*i, curr_data[i, :, 0], width=0.15, color=colors[i], label='factor %i' % tfs[i])
    ax.legend()
    axs.clean_axis(ax)
    ax.set_title('# of unitigs')
    ax.set_xticks(sp.arange(curr_data.shape[1]) + 0.3)
    ax.set_xticklabels(fbs)
    ax.set_xlabel('fallback')
    ### total length
    ax = fig.add_subplot(gs[0, 1])
    for i in range(len(tfs)):
        ax.bar(sp.arange(curr_data.shape[1]) + 0.15*i, curr_data[i, :, 1], width=0.15, color=colors[i], label='factor %i' % tfs[i])
    ax.legend()
    axs.clean_axis(ax)
    ax.set_title('Total length')
    ax.set_xticks(sp.arange(curr_data.shape[1]) + 0.3)
    ax.set_xticklabels(fbs)
    ax.set_xlabel('fallback')
    ### median unitig length
    ax = fig.add_subplot(gs[0, 2])
    for i in range(len(tfs)):
        ax.bar(sp.arange(curr_data.shape[1]) + 0.15*i, curr_data[i, :, 2], width=0.15, color=colors[i], label='factor %i' % tfs[i])
    ax.legend()
    axs.clean_axis(ax)
    ax.set_title('Unitig length (median)')
    ax.set_xticks(sp.arange(curr_data.shape[1]) + 0.3)
    ax.set_xticklabels(fbs)
    ax.set_xlabel('fallback')
    ### mean unitig length
    ax = fig.add_subplot(gs[1, 0])
    for i in range(len(tfs)):
        ax.bar(sp.arange(curr_data.shape[1]) + 0.15*i, curr_data[i, :, 3], width=0.15, color=colors[i], label='factor %i' % tfs[i])
    ax.legend()
    axs.clean_axis(ax)
    ax.set_title('Unitig length (mean)')
    ax.set_xticks(sp.arange(curr_data.shape[1]) + 0.3)
    ax.set_xticklabels(fbs)
    ax.set_xlabel('fallback')
    ### max unitig length
    ax = fig.add_subplot(gs[1, 1])
    for i in range(len(tfs)):
        ax.bar(sp.arange(curr_data.shape[1]) + 0.15*i, curr_data[i, :, 4], width=0.15, color=colors[i], label='factor %i' % tfs[i])
    ax.legend()
    axs.clean_axis(ax)
    ax.set_title('Unitig length (max)')
    ax.set_xticks(sp.arange(curr_data.shape[1]) + 0.3)
    ax.set_xticklabels(fbs)
    ax.set_xlabel('fallback')
    ### fraction aligned
    ax = fig.add_subplot(gs[1, 2])
    for i in range(len(tfs)):
        ax.bar(sp.arange(curr_data.shape[1]) + 0.15*i, curr_data[i, :, 5], width=0.15, color=colors[i], label='factor %i' % tfs[i])
    ax.legend()
    axs.clean_axis(ax)
    ax.set_title('Fraction re-aligned')
    ax.set_xticks(sp.arange(curr_data.shape[1]) + 0.3)
    ax.set_xticklabels(fbs)
    ax.set_xlabel('fallback')
    ### sup title and storing
    plt.suptitle(sample)
    plt.savefig(os.path.join(plotdir, '%s_pstudy.png' % sample), format='png', bbox_inches='tight')
    plt.show()
```
