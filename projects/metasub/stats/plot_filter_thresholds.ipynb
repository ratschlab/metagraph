```python
%matplotlib inline
import sys 
import os
import matplotlib
import matplotlib.pyplot as plt 
import scipy as sp 
plt.style.use('ggplot')
```

```python
basedir = '/cluster/work/grlab/projects/metagenome/data/metasub/graphs'
plotdir = os.path.join(basedir, 'plots')
if not os.path.exists(plotdir):
    os.makedirs(plotdir)
```

```python
k = 19
for k in [19, 41]:
    ### load data
    data = sp.loadtxt(os.path.join(basedir, 'output_k%i_cleaned.thresholds.txt' % k), dtype='str', delimiter=':')
    data = data[:, 2].astype('int')
    
    ### make plots
    fig = plt.figure(figsize=(10, 8), dpi=200)
    ax = fig.add_subplot(111)
    ax.hist(data, bins=sp.arange(data.max()))
    ax.set_title('Filter thresholds for k=%i' % k)
    ax.set_xticks(sp.arange(data.max()) + 0.5)
    ax.set_xticklabels(sp.arange(data.max()))
    plt.savefig(os.path.join(plotdir, 'filter_thresholds_k%i.png' % k), format='png', bbox_inches='tight')
    #plt.close(fig)
```
