```python
old_columns  = pd.read_table('~/big_graph/metasub/column_names.txt', header=None).values.ravel()
old_columns
```

```python
old_columns = pd.DataFrame({'old_name': old_columns})
```

```python
old_columns['local_id'] = old_columns.loc[:, 'old_name'].apply(lambda x: x.split('/')[-1].split('.k15.sequences.fasta.gz')[0])
```

```python
metadata = pd.read_table('/cluster/work/grlab/projects/metagenome/results/metasub_wasabi_complete_metadata_extended.csv', sep=',')
metadata
```

```python
metadata = metadata[~metadata['local_raw'].isnull()]
```

```python
metadata['local_id'] = metadata.loc[:, 'local_raw'].apply(lambda x: x.split('/')[-1].split('_2.fastq.gz')[0])
```

```python
merged = pd.merge(old_columns, metadata, on='local_id', how='inner')
merged
```

```python
merged = merged.replace(np.nan, '', regex=True)
```

```python
with open('rename_columns.txt', 'w') as f:
    for i in range(merged.shape[0]):
        meta_annotation = ";".join([
            str(merged.loc[i, 'uuid']),
            "MetasubName=" + str(merged.loc[i, 'metasub_name']),
            "City=" + str(merged.loc[i, 'city']),
            "latitude=" + str(merged.loc[i, 'latitude']),
            "longitude=" + str(merged.loc[i, 'longitude']),
            "surface_material="  + str(merged.loc[i, 'surface_material']),
            "station="  + str(merged.loc[i, 'station']),
            "num_reads="  + str(merged.loc[i, 'num_reads']),
            "city_latitude="  + str(merged.loc[i, 'city_latitude']),
            "city_longitude="  + str(merged.loc[i, 'city_longitude']),
            "city_total_population="  + str(merged.loc[i, 'city_total_population']),
            "continent="  + str(merged.loc[i, 'continent']),
            "sample_type="  + str(merged.loc[i, 'sample_type']),
        ])
        meta_annotation = meta_annotation.replace(' ', '_')
        print("{} ".format(merged.loc[i, 'old_name']) + meta_annotation, file=f)
```
