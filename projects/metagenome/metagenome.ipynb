```python
for col in df.columns:
    print('\n', col, ':')
    print(df.loc[:, col].unique())
```

```python
for col in df.columns:
    if (len(df.loc[:, col].unique()) < 10000):
        fig, ax = plt.subplots(figsize=(15, 5))
        ax.set_title(col)
        try:
            pd.value_counts(df.loc[:, col]).iloc[:100].plot.bar(ax=ax)
            plt.show()
        except:
            pass
```

```python
print(df.iloc[[1, 10**5, -1]].T.to_string())
```

```python
for strategy in ['WGS', 'AMPLICON']:
    fig, axes = plt.subplots(1, 2, figsize=(15, 4))
    df.loc[df.LibraryStrategy == strategy, 'LoadDate'].astype("datetime64").hist(ax=axes[0], bins=30)
    axes[0].set_title('LoadDate')
    df.loc[df.LibraryStrategy == strategy, 'ReleaseDate'].astype("datetime64").hist(ax=axes[1], bins=30)
    axes[1].set_title('ReleaseDate')
    plt.suptitle('Metagenome, {}'.format(strategy))
    plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 9))
filtered_df.plot.scatter('RealSize', 'Total Size, Mb', ax=ax)
ax.set_xscale('symlog')
ax.set_yscale('symlog')
ax.set_xlim([-1, 10**12])
ax.set_ylim([-1, 10**6])
ax.set_ylabel('Total Size, Mb', size=20)
ax.set_xlabel('Size on Disk, bytes', size=20)
for i in np.argsort(filtered_df['Total Size, Mb'] / filtered_df['RealSize'])[::-1][:5].values:
    ax.annotate(filtered_df.iloc[i].loc['Run'], xy=(filtered_df.iloc[i].loc['RealSize'],
                                                    filtered_df.iloc[i].loc['Total Size, Mb']))
plt.grid()
plt.show()
```
